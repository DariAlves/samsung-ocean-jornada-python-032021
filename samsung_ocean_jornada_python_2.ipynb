```
# Operadores: not
# O operador inverte o valor de um booleano

print('True', True)
print('False', False)
print('not True', not True)
print('not False', not False)
```

```
# Operadores: and

print('True and True:', True and True)
print('False and True:', False and True)
print('True and False:', True and False)
print('False and False:', False and False)
```

```
# Operadores: or

print('True or True:', True or True)
print('False or True:', False or True)
print('True or False:', True or False)
print('False or False:', False or False)
```

```
# Diferença entre texto e número

print('1 == 1:', 1 == 1) # True
print('1.0 == 1:', 1.0 == 1) # True
print('"1" == 1:', "1" == 1) # False
print('True == 1:', True == 1) # True
print('True == 2:', True == 2) # False
print('False == 0:', False == 0) # True
print('False == False:', False == False) #True

print()

print(type(1), type(1.0), type("1"))
```
