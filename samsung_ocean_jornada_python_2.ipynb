```
# Operadores: not
# O operador inverte o valor de um booleano

print('True', True)
print('False', False)
print('not True', not True)
print('not False', not False)
```

```
# Operadores: and

print('True and True:', True and True)
print('False and True:', False and True)
print('True and False:', True and False)
print('False and False:', False and False)
```

```
# Operadores: or

print('True or True:', True or True)
print('False or True:', False or True)
print('True or False:', True or False)
print('False or False:', False or False)
```

```
# Diferença entre texto e número

print('1 == 1:', 1 == 1) # True
print('1.0 == 1:', 1.0 == 1) # True
print('"1" == 1:', "1" == 1) # False
print('True == 1:', True == 1) # True
print('True == 2:', True == 2) # False
print('False == 0:', False == 0) # True
print('False == False:', False == False) #True

print()

print(type(1), type(1.0), type("1"))
```

```
# Declaração de condição: if, elif (else if), else
# IF: if (se), ou seja, se algo for True
# Indentation block
# Bloco indentado

if True:
  print('Esse if será lido')

if False:
  print('Esse if não será lido')
  
print('Apenas o primeiro será lido')
```

```
# Declaração de condição: if, elif (else if), else
# ELSE: else (caso contrário) e ele sempre vem depois de um if ou de um elif
# Portanto, ele sempre possui uma condição que foi checada antes dele
# O else é ativado caso a(s) condições anteriores foram False

if False:
  print('Esse if não será lido, pois a condição é False')
else:
  print('Esse else será lido')
```

```
# Declaração de condição: if, elif (else if), else
# ELIF: (caso contrário, se), ou seja, quando todas as condições anteriores foram False, mas sua condição é True
# 

if False:
  print('1 - Esse if não será lido, pois a condição é False')
elif False:
  print('2 - Esse elif não será lido')
elif True:
  print('3 - Esse elif será lido')
```

```
# Funções
# Para criar uma função usamos a declaração 'def', seguida de parênteses e criando um bloco com o símbolo ':'

def exibir_nome():
  nome = input('digite nome: ').upper()
  print(f'Bom dia, {nome}')

exibir_nome()
```

```
idade = 0
print('Idade fora da função:', idade)

def exibir_faixa_etaria():
  global idade 
  idade = 30

exibir_faixa_etaria()

faixa_etaria_1 = idade >= 0 
faixa_etaria_2 = idade >= 18
faixa_etaria_3 = idade >= 40

print('Idade global da função:', idade)
print('faixa_etaria_1:', faixa_etaria_1)
print('faixa_etaria_2:', faixa_etaria_2)
print('faixa_etaria_3:', faixa_etaria_3)

if faixa_etaria_3:
  print('Essa pessoa está na faixa etária 3')
elif faixa_etaria_2:
  print('Essa pessoa está na faixa etária 2')
elif faixa_etaria_1:
  print('Essa pessoa está na faixa etária 1')
else:
  print('Escreva uma idade correta!')
```

```
def exibir_faixa_etaria(idade):
  faixa_etaria_1 = idade >= 0 
  faixa_etaria_2 = idade >= 18
  faixa_etaria_3 = idade >= 40

  if faixa_etaria_3:
    print('Essa pessoa está na faixa etária 3')
  elif faixa_etaria_2:
    print('Essa pessoa está na faixa etária 2')
  elif faixa_etaria_1:
    print('Essa pessoa está na faixa etária 1')
  else:
    print('Escreva uma idade correta!')

exibir_faixa_etaria(0)
exibir_faixa_etaria(38)
exibir_faixa_etaria(50)
exibir_faixa_etaria(-50)
```
