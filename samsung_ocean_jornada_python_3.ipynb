# Exercício 13
Nome: Convertendo Celsius/Farenheit
Objetivo: Escrever duas funções de conversão, uma de graus celsius em farenheit e a outra que faça o contrário.
Dificuldade: Principiante

1 - Crie um aplicativo de conversão entre as temperaturas Celsius e Farenheit.

2 - Primeiro o usuário deve escolher se vai entrar com a temperatura em Célsius ou Farenheit, depois a conversão escolhida é realizada.

3 - Se C é a temperatura em Celsius e F em Farenheit, as fórmulas de conversão são:
F = (9 * C / 5) + 32
C = 5 * (F - 32) / 9


```
# Declaração
def converter_celsius_fahrenheit(celsius):
  F = (9 * celsius / 5) + 32
  print(f'{celsius}°C = {F}°F')
  

converter_celsius_fahrenheit(0)

def converter_fahrenheit_celsius(fahrenheit):
  C = 5 * (fahrenheit - 32) / 9
  print(f'{fahrenheit}°F = {C}°C')

converter_fahrenheit_celsius(32)
# Execução

```

```
# Deixando o usuário escolher o que quer fazer

escolha_usuario = input("Você quer converter para qual escala? ('C' para Celsius ou 'F' para Fahrenheit)\n").upper()

if escolha_usuario == "C":
  escala = "F"
elif escolha_usuario == "F":
  escala = "C"

temperatura = float(input(f"Digite o valor da temperatura, em º{escala}:"))

if escolha_usuario == "C":
  converter_fahrenheit_celsius(temperatura) 
elif escolha_usuario == "F":
  converter_celsius_fahrenheit(temperatura) 
```
